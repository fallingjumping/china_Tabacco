```python
import pandas as pd
import numpy as np
```

```python
file_name = './D线薄板烘丝的历史数据/20230102.xls'
data = pd.read_excel(file_name)
```

```python
deti = -380
data['new_afdry_mois']  = data['KLD烘后水分（%）d'].shift(periods=int(deti / 10), fill_value=-1)
data['new_afdry_temp']  = data['KLD烘后温度（℃）d'].shift(periods=int(deti / 10), fill_value=-1)
```

```python
from itertools import groupby
a = groupby(data['KLD排潮风门开度（%）d'].tolist())
j = 0
for i, v in a:
    if i == 50:
        print(len([_ for _ in v]))
```

```python
# print(len(data))
da = data # [4000:]
index1 = da[(da['烘前叶丝流量（kg/h）d'] != 0) & (da['new_afdry_mois'] > 0)].index.tolist()
index1[-10]

file_name = './mis_data/'
start = index1[0]
y = len(index1)-1
for i in range(y):
    end = 0
    if index1[i+1] - index1[i] > 1 or i == y - 1:
        end = index1[i]
        if end - start < 600:
            start = index1[i]
            continue
        new = data.iloc[start:end+2,:]

        a = groupby(new['KLD排潮风门开度（%）d'].tolist())
        l = 0
        for j, v in a:
            l -= len([_ for _ in v])
            if j == 50:
                df = data['KLD排潮风门开度（%）d'].shift(periods=l, fill_value=-1)[start:end+2]
                new['KLD排潮风门开度（%）d'] = df
                break
        # KLD排潮负压（μbar）d < 0
        
        new.to_csv(file_name + str(i) + '.csv',encoding='utf-16',sep='\t')
        start = index1[i+1]
```
